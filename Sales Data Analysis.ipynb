# Total Sales by Year

 Identify which year generated the highest revenue.

```python
sales_by_year = df.groupby('YEAR_ID')['SALES'].sum().reset_index()
sales_by_year
```

```python
import matplotlib.pyplot as plt

sales_by_year.plot(x='YEAR_ID', y='SALES', kind='bar')
plt.title('Total Sales by Year')
plt.ylabel('Sales')
plt.show()
```

# Monthly Sales Trend

Identify seasonal sales patterns.

```python
monthly_sales = df.groupby('MONTH_ID')['SALES'].sum().reset_index()
monthly_sales
```

```python
monthly_sales.plot(x='MONTH_ID', y='SALES', kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```
